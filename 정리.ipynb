```python
rate=1.5
for i,col in enumerate(cols):
    ## 각 col마다 iqr 범위
    quartile_1 = df[col].quantile(0.25) #1 사분위수 (25% 지점수)
    quartile_3 = df[col].quantile(0.75) #3 사분위수 (70% 지점수)
    IQR_range=abs(quartile_1-quartile_3)*rate
    IQR_under=quartile_1-IQR_range
    IQR_up=quartile_3+IQR_range
    
    
    
    outlier=np.zeros(len(df[col]))
 

    df_prime=df[col].values
    for j,value in enumerate(df_prime):
        if value>IQR_up or value<IQR_under:
            outlier[j]=1
            
   
    df_prime=pd.DataFrame(df_prime,columns=[col])
    outlier=pd.DataFrame(outlier)
    df_prime=df_prime.assign(outlier=outlier)

    fig=plt.figure(figsize=(20,20))
    ax=fig.add_subplot(8,4,i+1)
    fig.tight_layout()  
    plt.rcParams['figure.figsize'] = [10, 10]
    sns.scatterplot(df_prime[col],df_prime[col],hue='outlier',data=df_prime)
    
    
plt.savefig('scatter_IQR.png')
```

```python
rate=1.5
IQR_outliers_total=[]
for i,col in enumerate(cols):
    ## 각 col마다 iqr 범위
    quartile_1 = df[col].quantile(0.25) #1 사분위수 (25% 지점수)
    quartile_3 = df[col].quantile(0.75) #3 사분위수 (70% 지점수)
    IQR_range=abs(quartile_1-quartile_3)*rate
    IQR_under=quartile_1-IQR_range
    IQR_up=quartile_3+IQR_range
    
    count=0
    
    outlier=np.zeros(len(df[col]))
 

    df_prime=df[col].values
    for j,value in enumerate(df_prime):
        if value>IQR_up or value<IQR_under:
            outlier[j]=-1
   
    outlier=pd.DataFrame(outlier,columns=['outlier'])

    IQR_outlier=outlier.loc[outlier['outlier']==-1]
    temp=list(IQR_outlier.index)
    IQR_outliers_total.extend(temp)
```

## :isolation forest 기반으로 outlier detect하고, 그 index를 반환하는 함수 

```python

def if_outlier(df,x_name,y_name):
    from sklearn.ensemble import IsolationForest
    import pandas as pd
    df_prime=pd.DataFrame(df[x_name].values,columns=x_name)
    
    clf=IsolationForest(random_state=42)
    clf.fit(df_prime)
    y_predict=clf.predict(df_prime)
    df_prime=df_prime.assign(y=y_predict)
    
    display(df_prime.head())
    outliers=df_prime.loc[df_prime['y']==-1]
    index_outliers=list(outliers.index)
    return index_outliers
    
 
```

```python
name='Class'
x_name=list(df.columns)
output=if_outlier(df,x_name,name)
print(len(output))
```
