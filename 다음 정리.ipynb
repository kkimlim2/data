```python
## 라이브러리 import 하고 데이터셋 나누기(좀 더 빨리 fit시키기 위함)

PATH='C:/Users/98gud/Downloads/creditcard.csv/creditcard.csv'
import pandas as pd
df=pd.read_csv(PATH)
cols=df.columns
print(cols)
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style("darkgrid")
from sklearn.ensemble import IsolationForest

y_true=pd.DataFrame(df[['Class']].values)
df=df.drop(df[['Class']],axis=1)
display(y_true.head())
display(df.head())

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(df,y_true,test_size=0.3,shuffle=True,random_state=42)
```

```python
cols=list(cols)
del cols[len(cols)-1]
```

```python
rate=1.5
IQR_outliers_total=[]

for i,col in enumerate(cols):
    ## 각 col마다 iqr 범위
    quartile_1 = df[col].quantile(0.25) #1 사분위수 (25% 지점수)
    quartile_3 = df[col].quantile(0.75) #3 사분위수 (70% 지점수)
    IQR_range=abs(quartile_1-quartile_3)*rate
    IQR_under=quartile_1-IQR_range
    IQR_up=quartile_3+IQR_range
    
    count=0
    
    outlier=np.zeros(len(df[col]))
 

    X_train_prime=df[col].values
    for j,value in enumerate(X_train_prime):
        if value>IQR_up or value<IQR_under:
            outlier[j]=-1
   
    outlier=pd.DataFrame(outlier,columns=['outlier'])

    IQR_outlier=outlier.loc[outlier['outlier']==-1]
    temp=list(IQR_outlier.index)
    IQR_outliers_total.extend(temp)
```
